# Introduction

The agenda of writing this code is to generate a pencil sketch of any image. We will use the OpenCV library for this project.

Step 1: Find an image that you want to convert to a pencil sketch. We are going to use a flower image. You can choose whatever you want.

Step 2: Read the image in RBG format and then convert it to a grayscale image. Now, the image is turned into a classic black and white photo.

Step 3: Invert the grayscale image also called the negative image, this will be our inverted grayscale image. Inversion is basically used to enhance details.

Step 4: Finally create the pencil sketch by mixing the grayscale image with the inverted blurry image. This is done by dividing the grayscale image by the inverted blurry image.

Lastly, we will get our pencil_sketch. So, display it using OpenCV.

```python
!pip install opencv-python
```

```python
import cv2

# reading image
image = cv2.imread("Higanbana.jpg")

# converting BGR/color image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# image inversion
inverted_image = 255 - gray_image

# dividing grayscale image by the inverted blurry image
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)

# show the pencil sketch image
cv2.imshow("Original Image", image)
cv2.imshow("Pencil Sketch of Higanbana", pencil_sketch)
cv2.waitKey(0)
```
